```python
f = lambda x:x**2
df = lambda x: 2*x

def grad(f,gradf,x0,d,eta):
    
    x_values = [x0]
    func_values = [f(x0)]
    x1 = x0 -  eta*gradf(x0)
    itrns = 0
    
    while abs(x1-x0) > d or itrns<100 :
        x_values.append(x1)
        func_values.append(f(x1))
        x0 = x1
        x1 = x0 - eta*gradf(x0)
        itrns = itrns + 1
    
    plt.plot(x,f(x))
    plt.scatter(x_values,func_values)
    plt.xlabel('x', labelpad=20)
    plt.ylabel('y', labelpad=20)
    plt.title(f"Plotting using Gradient descent function")
    
grad(f,df,4,1e-5,0.1)
```
